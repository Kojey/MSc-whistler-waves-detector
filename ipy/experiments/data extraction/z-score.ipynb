```python
import sys
import os

sys.path.insert(0,'../..')
sys.path.insert(0,'../../py')
import parameters
import utilities
import spectrogram_utilities
import output_utilities
import spectrogram_output_visualiser

import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from scipy import stats
from sklearn import preprocessing
```

```python
# matplotlib.get_backend()
%matplotlib inline
```

```python
awd_event = 1
site = parameters.sites[0]
```

```python
# Select random file
file = utilities.random_file(awd_event, site)
# file = '2013-06-04UT17:37:09.15700602.marion.vr2'
# file = '2013-06-04UT17:37:09.15700602.marion.vr2'
file = '2013-05-28UT14:08:19.75372164.marion.vr2'
indices, times, frequencies, spectrogram = spectrogram_output_visualiser.spectrogram_and_output(awd_event, site, file, output=False, verbose=True)

fig = plt.figure(figsize=(16,5))
img = plt.pcolormesh(times,frequencies/1000,spectrogram, cmap='jet')
fig.colorbar(mappable=img, label='Density [V**2/Hz]')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (kHz)')
plt.title(file)
plt.show()

time = [510, 3420, 2600] # ms
time_res = 4.8 # in ms
fig = plt.figure(figsize=(16,5))
for t in time:
    index = int(t/time_res)
    plt.plot(frequencies/1000, spectrogram[:,index])
plt.xlabel('Frequency (kHz)')
plt.ylabel('Z-score')
plt.title('Time cuts of the spectrogram')
plt.legend(['No spherics','Spherics','Whistler'])
plt.show()


_spectrogram = stats.zscore(spectrogram, axis=0)
fig = plt.figure(figsize=(16,5))
for t in time:
    index = int(t/time_res)
    plt.plot(frequencies/1000, _spectrogram[:,index])
plt.xlabel('Frequency (kHz)')
plt.ylabel('Z-score')
plt.title('Time cuts of Z-score performance on the frequency axis')
plt.legend(['No spherics','Spherics','Whistler'])
plt.show()

_spectrogram = stats.zscore(_spectrogram, axis=1)
_spectrogram = stats.zscore(_spectrogram, axis=0)
fig = plt.figure(figsize=(16,5))
for t in time:
    index = int(t/time_res)
    plt.plot(frequencies/1000, _spectrogram[:,index])
plt.xlabel('Frequency (kHz)')
plt.ylabel('Z-score')
plt.title('Time cuts of Z-score performance on the time and frequency axis')
plt.legend(['No spherics','Spherics','Whistler'])
plt.show()

fig = plt.figure(figsize=(16,5))
img = plt.pcolormesh(times,frequencies/1000,_spectrogram, cmap='jet')
fig.colorbar(mappable=img, label='Density [V**2/Hz]')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (kHz)')
plt.title(file)
plt.show()
```
